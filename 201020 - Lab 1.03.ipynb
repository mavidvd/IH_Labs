```python
df1 = df1[['Customer', 'State', 'Gender', 'Education', 'Cust Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
df2 = df2[['Customer', 'State', 'Gender', 'Education', 'Cust Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
df3 = df3[['Customer', 'State', 'Gender', 'Education', 'Cust Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
data = data.dropna(axis = 0)
```

```python
data
```

```python
def remove_pct(x):
    x = str(x)
    if x.endswith('%'):
        return round(float(x[:-1])/100, 2)
    else:
        return round(float(x), 2)
```

```python
data['Cust Lifetime Value'] = data['Cust Lifetime Value'].apply(remove_pct)
```
