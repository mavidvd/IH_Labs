```python
rentals_may = pd.read_sql('SELECT * FROM rental where MONTH(rental_date)=5', engine)
```

```python
rentals_june = pd.read_sql('SELECT * FROM rental where MONTH(rental_date)=6', engine)
```

```python
rentals_may2 = pd.read_sql('SELECT customer_id, count(*) FROM rental where MONTH(rental_date)=5 group by customer_id', engine)
```

```python
rentals_may2
```

```python
rentals_june2 = pd.read_sql('SELECT customer_id, count(*) FROM rental where MONTH(rental_date)=6 group by customer_id', engine)
```

```python
rentals_june2
```

```python
sum(rentals_june2['count(*)'])
```

```python
rentals_mj = pd.merge(rentals_may2, rentals_june2, on = 'customer_id', suffixes = ('_may', '_june'))
```

```python
rentals_mj
```

```python
conditions = [
    (rentals_mj['count(*)_may'] > rentals_mj['count(*)_june']),
    (rentals_mj['count(*)_may'] < rentals_mj['count(*)_june']),
    (rentals_mj['count(*)_may'] == rentals_mj['count(*)_june'])]
```

```python
values = ['May', 'June', 'Same in May and June']
```

```python
rentals_mj['Month with most rentals'] = np.select(conditions, values)
```
