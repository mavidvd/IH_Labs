```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from tqdm.notebook import tqdm
```

```python
url = requests.get('https://www.billboard.com/charts/hot-100')
```

```python
s_song = BeautifulSoup(url.text, 'html.parser')
```

```python
song_list = s_song.find_all('button', 'chart-element__wrapper display--flex flex--grow sort--default')
```

```python
songs = []

for song in song_list:
    rank = song.find('span', 'chart-element__rank__number')
    name = song.find('span', 'chart-element__information__song text--truncate color--primary')
    artist = song.find('span', 'chart-element__information__artist text--truncate color--secondary')
    songs.append([rank.text, name.text, artist.text])
```

```python
songs
```

```python
df = pd.DataFrame.from_records(songs, columns = ['Rank', 'Song', 'Artist'], index = 'Rank')
```
